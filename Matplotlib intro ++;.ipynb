```python
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt #same as of the first line
import numpy as np
```

```python
x = np.array([-3,-2,-1,0,1,2,3])
x
```

```python
y = x**2
y
```

```python
# now plotting this
plt.plot(x,y)
plt.show()
# aik parabola aagyi hai cause square of x ka graph hai yeh
```

```python
# sub plots -> iska mtlb hai ager hum aik hi figure k sub plots bana chahain
fig,axs = plt.subplots() #iske arguments main (NoOfRows , NoOfColumns) yani aik hi figure main kitne graphs chahiyan. by default it is 1,1
axs.plot(x,y)
plt.show()
```

```python
xd,axs = plt.subplots() #xd is figure ka name hai yahan isko mention krna zaruri hai 
axs.plot(x,y)
plt.show()
```

```python
figure,axes = plt.subplots(2,2) #2,2 is the shape of figure
axes[0,0].plot(x,y)
axes[0,1].plot(x,x ** 3)
axes[1,0].plot(x,x ** 4)
axes[1,1].plot(x,x ** 3)
plt.show()
```
