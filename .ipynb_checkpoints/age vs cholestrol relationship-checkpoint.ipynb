```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart_disease_uci.csv")
```

```python
df
```

```python
# now starting visualizing the graph
# let us plot the relationship between age and cholestrol level

fig,axs = plt.subplots()
# axs.scatter(df["age"].values,df["chol"].values) #values krne se yeh numpy values main change hogye

axs.scatter(df["age"],df["chol"]) #although as a df of pandas it is also valid
plt.show()

# As the graph is not symmetric we can conclude that there isn't any clear relation in between them
```

```python


fig,axs = plt.subplots()
axs.scatter(df["age"].values[:10],df["chol"].values[:10]) #values krne se yeh numpy values main change hogye

plt.show()

# shuru ki 10 values lekar hisaab lagaya gya hai!
```

```python
# now if we want to plot for the mean of ages  vs their cholestrol level
df_meanByAges = df.groupby("age").mean()
df_meanByAges
# ab har individuakl ki bjayee ages k mean lekar cholestrol , etc ka average lia gya hai
```

```python
fig,axs = plt.subplots()
axs.scatter(df_meanByAges.index , df_meanByAges.chol) #values krne se yeh numpy values main change hogye

plt.show()

# The insight im getting is with the passage of time cholestrol barh rha hai!!
```
