```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart_disease_uci.csv")
df_groupByAge = df.groupby("age").mean()
```

```python
# LINE Graph
fig,axes = plt.subplots()
axes.plot(df_groupByAge.index,df_groupByAge.chol)
plt.show()
```

```python
# Scatter Graph
fig,axes = plt.subplots()
axes.scatter(df_groupByAge.index,df_groupByAge.chol)
plt.show()
```

```python
# Bar Graph 
fig,axes = plt.subplots()
axes.bar(df_groupByAge.index,df_groupByAge.chol)
plt.show()
```

```python
# Histogram 
fig,axes = plt.subplots()
axes.hist(df["age"])
plt.show()
```
